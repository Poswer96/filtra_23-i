# 3. Filtros Pasa Altas Tipo Butterworth

```python
#NOTE importar bibliotecas
import numpy             as np
import matplotlib.pyplot as plt

from scipy import signal
```

```python
s = [-1., 0.]

plt.axvspan(s[0],s[1], facecolor='b', alpha=0.5)
plt.xticks( -np.arange( 5  ) )
plt.yticks(  np.arange(-5,5) )
plt.grid(True)
plt.title('Plano S')
plt.xlabel(u'$\mathbb{R}e$',fontsize=16)
plt.ylabel(u'$\mathbb{I}m$',fontsize=16)
plt.show()
```

```python
r = -1.5
s = np.linspace(r,-.1,1000)

plt.plot(s ,-1/s, '-b' ,linewidth=3)
plt.plot(-1, 1  , 'og' ,markersize=20)
plt.axvspan(r-1.99,-1., ymin=0  , ymax=.1, alpha=0.5, color='red')
plt.axvspan(-1    ,-.01, ymin=.1, ymax=1., alpha=0.5, color='green')
plt.grid(True)
plt.title('S vs -1/S')
plt.xlabel(u'S'  ,fontsize=16)
plt.ylabel(u'1/S',fontsize=16)
plt.show()
```
